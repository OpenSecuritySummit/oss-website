```python
%autoreload

hugo   = API_Hugo_OSS()
#hugo.df_sessions(type='track')
#hugo.df_sessions(type='working-session')['participants']
mapping = []
for path,session in hugo.sessions().items():
    content      = session.get('content')
    metadata     = session.get('metadata')
    page_type    = metadata.get('type')
    title        = metadata.get('title')
    track        = metadata.get('track')
    organizers   = metadata.get('organizers')
    participants = metadata.get('participants')
    if not organizers: organizers = []
    if not participants: participants = []
    if type(organizers)   is str: organizers   = organizers.split(',')
    if type(participants) is str: participants = participants.split(',')
    if 'TBD' in organizers: organizers.remove('TBD')
    if 'Pending' in organizers: organizers.remove('Pending')
        
    if title and page_type:
        item = { 
                    'title'       : title            , 
                    'track'       : track            ,             
                    'page_type'   : page_type        , 
                    'organizers'  : organizers       , 
                    'participants': participants     , 
                    'content'     : len(content)     , 
                    'path'        : path  
        }
        mapping.append(item)

df_mapping = pd.DataFrame(mapping)
df_mapping
```

```python
df_mapping  = df_mapping[['title', 'track', 'page_type' , 'content', 'organizers', 'participants']]
df_sessions = df_mapping[df_mapping['page_type'] != 'track']
#df_sessions['content'] = df_mapping['content'].apply(lambda x:len(x))
#df_sessions['participants'].fillna('')
df_sessions = df_sessions.sort_values(['content'],ascending=False).reset_index(drop=True).set_index('title')


df_sessions
```
